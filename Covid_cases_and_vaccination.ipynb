```python
import pandas as pd
%matplotlib inline
```

```python
url="https://covid.ourworldindata.org/data/owid-covid-data.csv"
df_country = pd.read_csv(url, sep=",")
df_country.columns
```

```python
df_country['location'].unique()
```

```python
df_India=df_country[df_country['location']=="India"]
df_India.head()
```

```python
df_India=df_India.reset_index(drop=True)
df_India.head()
```

# Plotting Chart of Covid Cases

```python
import plotly.express as px
fig=px.line(df_India, x="date", y="total_cases", title="Covid Cases")
fig.show()
```

```python
b_India = df_country['location']=="India"
b_Germany = df_country['location']=="Germany"
b_Italy = df_country['location']=="Italy"
```

```python
b_all = b_India|b_Germany|b_Italy
df_my_list = df_country[b_all]
```

```python
fig=px.line(df_my_list, x="date", y="total_cases", color = "location", title="Covid Cases")
fig.show()
```

```python

df_my_list['total_normal_cases']=df_my_list['total_cases']/df_my_list['population']
fig=px.line(df_my_list, x="date", y="total_normal_cases", color = "location", title="Covid Cases")
fig.update_layout(
    title="Relative cases overtime of Covid infectors",
    xaxis_title="Time",
    yaxis_title="Number of Cases",
    legend_title="Country",
    plot_bgcolor="#fdf9ed",
    font=dict(
        family="verdana",
        size=10,
        color="black"
    )
)
# fig.write_image("Covid_Cases.png")
fig.show()
```
